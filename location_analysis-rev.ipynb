```python
from pandas.compat import StringIO
from collections import Counter
import pandas
```

```python
pro_df = pandas.read_csv('data_profil.csv')
pro_df
```

```python

csv_string: str = open('catatan_lokasi.csv').read()
csv_string = csv_string.replace('/19', '/2019')
csv_string

str_io = StringIO(csv_string)
df_lokasi = pandas.read_csv(str_io)
df_lokasi['tanggal'] = pandas.to_datetime(df_lokasi['tanggal'], format = "%d/%m/%Y")
df_lokasi = df_lokasi.sort_values(by=["id", "tanggal"])
df_lokasi
```

```python
id_before = -1
start_loc = []
end_loc = []

for row in df_lokasi.iterrows():
    current_id = row[1]['id']
    location = row[1]['lokasi_dominan']
    
    if (current_id != id_before):
        start_loc.append(location)
        end_loc.append(location)
    else :
        end_loc[-1] = location
    id_before = current_id

```

```python
all([start_loc[i] == end_loc[i] for i in range(len(start_loc))])
```

```python
join_df = pro_df.join(df_lokasi.set_index('id'), on ="id")
#join_df.to_csv('data_join.csv', index=False)
join_df
```

```python
kota_mudik = []
for i, row, in enumerate(join_df.iterrows()):
    if (i % 26 == 15):
        kota_mudik.append(row[1]['lokasi_dominan'])
kota_mudik
```
